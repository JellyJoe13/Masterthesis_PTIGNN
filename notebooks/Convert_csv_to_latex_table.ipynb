# Convert csv to latex table

## Imports

```python
import pandas as pd
```

## Loading dataframe

```python
file_path = "hyperoptimization/results/tox21_chienn/final/rank1_99da0fb6_results.csv"
```

```python
df = pd.read_csv(file_path)
display(df)
```

```python
print(df.to_latex())
```

```python
def custom_convert(df, index_col_name: str = "epoch"):
    # get data
    n_rows = len(df)
    n_cols = len(df.iloc[0]) + 1
    
    # init return string
    output_str = "\\begin{table}\n\t\\centering\n\t\\begin{tabular}{"

    def custom_merger(l):
        o = "|"
        for e in l:
            o += e + "|"
        return o

    output_str += f"{custom_merger(['c' for _ in range(n_cols)])}"
    output_str += "}\n"
    output_str += "\t\t\\hline"

    # data area
    prefix = "\n\t\t"

    # header
    output_str += prefix + index_col_name
    for col in df.columns:
        t = col.replace('_', '\\_')
        output_str += f" & {t}"
    output_str += "\\\\" + prefix + "\\hline\\hline"

    # body
    for idx, row in df.iterrows():
        output_str += prefix + f"{idx}"
        for col in row:
            output_str += f" & {col}"
        output_str += "\\\\" + prefix + "\\hline"

    # closing area
    output_str += "\n\t\\end{tabular}\n\\end{table}"

    return output_str
```

```python
print(custom_convert(df))
```
